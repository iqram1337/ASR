```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import signal

plt.style.use("bmh")
```

```python
def saladass(d=5, N=6):
    #w,h = signal.freqz(d,N)
    k = np.linspace(0, 0.2, 1000)

    amp_resp = 20*np.log(abs((np.sin(np.pi*d*N*k))/(N*np.sin(np.pi*d*k))))

    plt.plot(k, amp_resp)

    plt.ylim(bottom=-100, top=0)

saladass()
saladass(N=3)
plt.show()
```

```python
def plot_freqz(b,a=1):
    w,h = signal.freqz(b,a)
    #print(w,h)
    h_dB = 20 * np.log10 (abs(h))
    plt.plot(w/np.pi, h_dB)
    plt.ylim([max(min(h_dB), -100) , 5])
    

b = np.repeat(1.0/11, 11)

plot_freqz(b)
```
